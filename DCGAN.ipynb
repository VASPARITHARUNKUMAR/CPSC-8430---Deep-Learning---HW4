```python
class Discriminator(nn.Module):
    def __init__(self, channels_img, features_d):
        super(Discriminator, self).__init__()
        self.disc = nn.Sequential(
            # input: N x channels_img x 64 x 64
            nn.Conv2d(
                channels_img, features_d, kernel_size=4, stride=2, padding=1
            ),
            nn.LeakyReLU(0.2),
            # _block(in_channels, out_channels, kernel_size, stride, padding)
            self.Dnet(features_d, features_d * 2, 4, 2, 1),      #out=128,k=5,s=2,p=1
            self.Dnet(features_d * 2, features_d * 4, 4, 2, 1),  #out=256,k=5,s=2,p=1
            self.Dnet(features_d * 4, features_d * 8, 4, 2, 1),  #out=512,k=5,s=2,p=1
            # After all _block img output is 4x4 (Conv2d below makes into 1x1)
            nn.Conv2d(features_d * 8, 1, kernel_size=4, stride=2, padding=0),
            nn.Sigmoid(),
        )

    def Dnet(self, in_channels, out_channels, kernel_size, stride, padding):
        return nn.Sequential(
            nn.Conv2d(
                in_channels,
                out_channels,
                kernel_size,
                stride,
                padding,
                bias=False,
            ),
            #nn.BatchNorm2d(out_channels),
            nn.LeakyReLU(0.2),
        )

    def forward(self, x):
        return self.disc(x)
```

```python
class Generator(nn.Module):
    def __init__(self, channels_noise, channels_img, features_g):
        super(Generator, self).__init__()
        self.net = nn.Sequential(
            # Input: N x channels_noise x 1 x 1
            #Gnet(in_channels, out_channels, kernel_size, stride, padding)
            self.Gnet(channels_noise, features_g * 16, 4, 1, 0),  # img: 4x4
            self.Gnet(features_g * 16, features_g * 8, 4, 2, 1),  # img: 8x8
            self.Gnet(features_g * 8, features_g * 4, 4, 2, 1),  # img: 16x16
            self.Gnet(features_g * 4, features_g * 2, 4, 2, 1),  # img: 32x32
            nn.ConvTranspose2d(
                features_g * 2, channels_img, kernel_size=4, stride=2, padding=1
            ),
            # Output: N x channels_img x 64 x 64
            nn.Tanh(),
        )

    def Gnet(self, in_channels, out_channels, kernel_size, stride, padding):
        return nn.Sequential(
            nn.ConvTranspose2d(
                in_channels,
                out_channels,
                kernel_size,
                stride,
                padding,
                bias=False,
            ),
            nn.BatchNorm2d(out_channels,momentum=0.9),
            nn.ReLU(),
        )

    def forward(self, x):
        return self.net(x)
```

```python
def weight_initializer(model):
    # Initializes weights according to the DCGAN paper
    for m in model.modules():
        if isinstance(m, (nn.Conv2d, nn.ConvTranspose2d, nn.BatchNorm2d)):
            nn.init.normal_(m.weight.data, 0.0, 0.02)

#Intitating instances
gen = Generator(NOISE_DIM, CHANNELS_IMG, FEATURES_GEN).to(device)
disc = Discriminator(CHANNELS_IMG, FEATURES_DISC).to(device)
weight_initializer(gen)
weight_initializer(disc)

```

```python
class InceptionV3Model(nn.Module):
    """Pretrained InceptionV3Model network returning feature maps"""

    # Index of default block of inception to return,
    # corresponds to output of final average pooling
    DEFAULT_FEATURE_BLOCK_INDEX = 3

    # Maps feature dimensionality to their output blocks indices
    FEATURE_BLOCK_INDEX_BY_DIM = {
        64: 0,   # First max pooling features
        192: 1,  # Second max pooling featurs
        768: 2,  # Pre-aux classifier features
        2048: 3  # Final average pooling features
    }

    def __init__(self,
                 return_blocks=[DEFAULT_FEATURE_BLOCK_INDEX],
                 adjust_input=True,
                 standardize_input=True,
                 allow_grad=False):
        
        super(InceptionV3Model, self).__init__()

        self.adjust_input = adjust_input
        self.standardize_input = standardize_input
        self.return_blocks = sorted(return_blocks)
        self.max_block_needed = max(return_blocks)

        assert self.max_block_needed <= 3, \
            'Last possible output block index is 3'

        self.blocks = nn.ModuleList()

        
        inception = models.inception_v3(pretrained=True)

        # Block 0: input to maxpool1
        block0 = [
            inception.Conv2d_1a_3x3,
            inception.Conv2d_2a_3x3,
            inception.Conv2d_2b_3x3,
            nn.MaxPool2d(kernel_size=3, stride=2)
        ]
        self.blocks.append(nn.Sequential(*block0))

        # Block 1: maxpool1 to maxpool2
        if self.max_block_needed >= 1:
            block1 = [
                inception.Conv2d_3b_1x1,
                inception.Conv2d_4a_3x3,
                nn.MaxPool2d(kernel_size=3, stride=2)
            ]
            self.blocks.append(nn.Sequential(*block1))

        # Block 2: maxpool2 to aux classifier
        if self.max_block_needed >= 2:
            block2 = [
                inception.Mixed_5b,
                inception.Mixed_5c,
                inception.Mixed_5d,
                inception.Mixed_6a,
                inception.Mixed_6b,
                inception.Mixed_6c,
                inception.Mixed_6d,
                inception.Mixed_6e,
            ]
            self.blocks.append(nn.Sequential(*block2))

        # Block 3: aux classifier to final avgpool
        if self.max_block_needed >= 3:
            block3 = [
                inception.Mixed_7a,
                inception.Mixed_7b,
                inception.Mixed_7c,
                nn.AdaptiveAvgPool2d(output_size=(1, 1))
            ]
            self.blocks.append(nn.Sequential(*block3))

        for param in self.parameters():
            param.allow_grad = allow_grad

    def forward(self, inp):
        outp = []
        x = inp

        if self.adjust_input:
            x = F.interpolate(x,
                              size=(299, 299),
                              mode='bilinear',
                              align_corners=False)

        if self.standardize_input:
            x = 2 * x - 1  # Scale from range (0, 1) to range (-1, 1)

        for idx, block in enumerate(self.blocks):
            x = block(x)
            if idx in self.return_blocks:
                outp.append(x)

            if idx == self.max_block_needed:
                break

        return outp
    
block_idx = InceptionV3Model.FEATURE_BLOCK_INDEX_BY_DIM[2048]
model = InceptionV3Model([block_idx])
model=model.cuda()
```

```python
def matrix_square_root_approx(A, method='ss'):
    if method == 'ss':
        # Scaling and squaring method
        U, s, V = np.linalg.svd(A)
        s = np.diag(np.sqrt(s))
        return U.dot(s).dot(V)
    elif method == 'pi':
        # Power iteration method
        max_iter = 100
        tol = 1e-6
        X = np.eye(A.shape[0])
        for i in range(max_iter):
            Y = 0.5 * (X + A.dot(np.linalg.inv(X)))
            if np.max(np.abs(X - Y)) < tol:
                break
            X = Y
        return X
    else:
        raise ValueError("Invalid method: %s" % method)
            
def compute_frechet_distance(mu_real, cov1, mu_fake, cov2, eps=1e-6):
    mu_real = np.atleast_1d(mu_real)
    mu_fake = np.atleast_1d(mu_fake)

    cov1 = np.atleast_2d(cov1)
    cov2 = np.atleast_2d(cov2)

    assert mu_real.shape == mu_fake.shape, \
        'Training and test mean vectors have different lengths'
    assert cov1.shape == cov2.shape, \
        'Training and test covariances have different dimensions'

    diff = mu_real - mu_fake
    
    cov_sqrt_mean = matrix_square_root_approx(cov1.dot(cov2), 'ss')

    if not np.isfinite(cov_sqrt_mean).all():
        msg = ('fid calculation produces singular product; '
               'adding %s to diagonal of cov estimates') % eps

        offset = np.eye(cov1.shape[0]) * eps
        cov_sqrt_mean = matrix_square_root_approx((cov1 + offset).dot(cov2 + offset))

    
    if np.iscomplexobj(cov_sqrt_mean):
        if not np.allclose(np.diagonal(cov_sqrt_mean).imag, 0, atol=1e-3):
            m = np.max(np.abs(cov_sqrt_mean.imag))
            raise ValueError('Imaginary component {}'.format(m))
        cov_sqrt_mean = cov_sqrt_mean.real
    trace_cov_sqrt_mean = np.trace(cov_sqrt_mean)

    return (diff.dot(diff) + np.trace(cov1) +
            np.trace(cov2) - 2 * trace_cov_sqrt_mean)
```

```python


def compute_frechet_inception_distance(images_real,images_fake,model):
    mu_1,std_real=compute_activation_statistics(images_real,model,cuda=True)
    mu_2,std_fake=compute_activation_statistics(images_fake,model,cuda=True)
    frechet_distance = compute_frechet_distance(mu_1, std_real, mu_2, std_fake)
    return frechet_distance

```

```python
gen_loss_values = []
disc_loss_values = []
generated_images = []
fid_scores = []
iteration_count = 0

print("Training Started")
# start timing
start_time = time.time() 
for epoch in range(NUM_EPOCHS):
    epoch = epoch+1
    
#     for batch_idx, (real, _) in enumerate(dataloader):

    for batch_idx, data in enumerate(dataloader,0):
        #training with real batch
        real = data[0].to(device)
        noise = torch.randn(BATCH_SIZE, NOISE_DIM, 1, 1).to(device)
        fake = gen(noise) #genrator img data

        ### Train Discriminator: max log(D(x)) + log(1 - D(G(z)))
        disc_real = disc(real).reshape(-1)
        disc_real_loss = criterion(disc_real, torch.ones_like(disc_real))
        disc_fake = disc(fake.detach()).reshape(-1)
        disc_fake_loss = criterion(disc_fake, torch.zeros_like(disc_fake))
        total_disc_loss = (disc_real_loss + disc_fake_loss) / 2
        disc.zero_grad()
        total_disc_loss.backward()
        optimDiscriminator.step()

        ### Train Generator: min log(1 - D(G(z))) <-> max log(D(G(z))
        output = disc(fake).reshape(-1)
        gen_loss = criterion(output, torch.ones_like(output))
        gen.zero_grad()
        gen_loss.backward()
        optimGenerator.step()
        gen_loss_values.append(gen_loss.detach().cpu())
        disc_loss_values.append(total_disc_loss.detach().cpu())
        
        
        # Check how the generator is doing by saving G's output on fixed_noise
        if (iteration_count % 500 == 0) or ((epoch == NUM_EPOCHS) and (batch_idx == len(dataloader)-1)):
            with torch.no_grad():
                fake = gen(fixed_noise).detach().cpu()
            generated_images.append(utils.make_grid(fake, padding=2, normalize=True))
            
        iteration_count += 1
    
    #Cal. FID
    fid_value=compute_frechet_inception_distance(real,fake,model)
    fid_scores.append(fid_value)
    
    if epoch%5 == 0:
        print( f'\nEpoch [{epoch}/{NUM_EPOCHS}]Batch {batch_idx}/{len(dataloader)} \
                  Loss Discriminator: {total_disc_loss:.3f}, loss Generator: {gen_loss:.3f} FID:{fid_value:.3f} ') #,end = '\r', flush=True)
    
    #Saving Generated image/epoch     
    images = gen(fixed_noise)
    if not os.path.exists('Results/DCGAN_FAKE'):
       os.makedirs('Results/DCGAN_FAKE')
    if not os.path.exists('Results/DCGAN_REAL'):
       os.makedirs('Results/DCGAN_REAL')
    utils.save_image(images.detach(),'Results/DCGAN_FAKE/DCGAN50_epoch_%03d.png' % (epoch), normalize = True)
    
    #Saving Real image/epoch 
#     real = data[0].to(device)
    utils.save_image(real.detach(),'Results/DCGAN_REAL/DCGAN50_epoch_%03d.png' % (epoch), normalize = True)
        
    
end_time = time.time()
print('\n\n elapsed time%.2f s.'%(end_time-start_time))
```
